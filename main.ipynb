```python
import vehicule
import ligne
import matplotlib.pyplot as plt

l = ligne.Ligne("profils/TRAM-Le_Havre_ligne_B.json")
veh = vehicule.Vehicule("vehicules/citadis_302.json")
```

```python
sol = l.profil_vitesse_acceleration(veh, 0, 0, True)

fig = plt.figure(1, dpi=300)
plt.title("Vitesse du {}")
ax = plt.gca()
ax.plot(sol.y[1] / 1000, 3.6 * sol.y[0])
ax.set_xlabel("distance (km)")
ax.set_ylabel("vitesse (km/h)")
plt.show()
```
